```python
import balokBeton.Balok as bk
import balokBeton.Sketch as sk
```

# Analysis Flanged

```python
rebar = [["D29",4]]
b1 = bk.DoublyFlanged()
b1.material()
b1.dimension(bw=300,h=600,be=900,hf=120)
b1.rebar(rebar=rebar,Dc="D22",nc=2)
b1.run()
```

```python
sk.Draw(b1).plot(size=(15,15),font=13)
```

```python
print("Moment resistance =",b1.Mn,"kNm")
print("Factored moment resistance =",b1.phiMn,"kNm")
```
