# Design

```python
b_mid = bk.Rect(name="Beam Design Result")
b_mid.material(fc=30,fy=400)
b_mid.dimension(bw=400, h=600, ds=40)
b_mid.design(Mu=180,D="D25")
```

```python
sk.Draw(b_mid).plot()
```

```python
sk.Draw(b_mid).plot(size=(10,10),font=10,stirrups=True,diagram=True)
```

# Analysis

```python
b_mid = bk.Rect(name="Beam Analysis Results")
b_mid.material(fc=30,fy=400)
b_mid.dimension(bw=400, h=600, ds=40.5)
b_mid.rebar(rebar = [["D25",2]], Dv="D10")
b_mid.run()
```

```python
sk.Draw(b_mid).plot()
```

# Display results

```python
print("Moment resistance =",b_mid.Mn,"kNm")
print("Factored moment resistance =",b_mid.phiMn,"kNm")
```
