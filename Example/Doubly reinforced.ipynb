```python
import balokBeton.Balok as bk
import balokBeton.Sketch as sk
```

# Analysis of Doubly Reinforced Beam

```python
b1 = bk.Doubly(name="Balok & Sloof")
b1.material(fc=21,fy=400)
b1.dimension(bw=250, h=400, ds=30)
b1.rebar(rebar = [["D16",3]], Dv="D10",Dc="D16",nc=2)
b1.run()
b1.shear(Vu=0)
```

```python
sk.Draw(b1).plot(size=(10,10),font=10,stirrups=True,diagram=True)
```

```python
print("Moment resistance =",b1.Mn,"kNm")
print("Factored moment resistance =",b1.phiMn,"kNm")
```

# Design Doubly Reinforced

```python
b2 = bk.Doubly()
b2.material()
b2.dimension(bw=300,h=600)
b2.design(Mu=250,D="D25",Dc="D19")
```

```python
sk.Draw(b2).plot(size=(10,10),font=10,stirrups=True,diagram=True)
```

```python
print("Moment resistance =",b2.Mn,"kNm")
print("Factored moment resistance =",b2.phiMn,"kNm")
```
